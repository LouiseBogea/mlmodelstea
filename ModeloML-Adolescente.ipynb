## Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import (
    f1_score
)
```

## Importando o conjunto de dados

```python
dataset = pd.read_csv('D:/Transdimension/Order 81004901 Comparison of Machine Learning Models/csv_result-Autism-Adolescent-Data.csv')
dataset.pop('ethnicity')
dataset.pop('id')
dataset.pop('used_app_before')
dataset.pop('result')
dataset.pop('age_desc')
dataset.pop('relation')
dataset.pop('contry_of_res')
```

```python
pd.DataFrame(dataset)
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

pd.DataFrame(X)
```

## One Hot Encoding conjunto de dados sobre pacientes

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(categories = 'auto'), [11,12,13])], remainder = 'passthrough')
X = np.array(ct.fit_transform(X))
```

```python
pd.DataFrame(X)
```

## One Hot Encoding no conjunto de dados de resultado

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
###
pd.DataFrame(y)
```

## Dividindo os dados em conjuntos de treino e teste

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Escalonando os dados

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, 16:] = sc.fit_transform(X_train[:, 16:])
X_test[:, 16:] = sc.transform(X_test[:, 16:])
```

```python
pd.DataFrame(X_train)
```

```python
pd.DataFrame(X_test)
```

# Modelos de classificaÃ§Ã£o:

# Random Forest

## Treinando o modelo

```python
## Random Forest Classification
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=190, criterion = 'entropy', random_state=0)
classifier.fit(X_train, y_train)
```

## Fazendo os resultados do modelo

```python
y_pred = classifier.predict(X_test)
#print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Matriz de confusÃ£o

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print("Matriz de confusÃ£o para Random Forest Classification:")
print(cm)
precisao_teste = accuracy_score(y_test, y_pred)
precisao_treino = accuracy_score(y_train, classifier.predict(X_train))
print("Precisao do treino e: ", precisao_treino)
print("Precisao do teste e: ", precisao_teste)
```

```python
f1score = f1_score(y_test, y_pred)
f1score
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
#
# Create an instance of Pipeline
#
pipeline = make_pipeline(StandardScaler(), RandomForestClassifier(n_estimators=100, max_depth=4))
#
# Pass instance of pipeline and training and test data set
# cv=10 represents the StratifiedKFold with 10 folds
#
scores = cross_val_score(pipeline, X=X_train, y=y_train, cv=5, n_jobs=1)
 
print('Cross Validation accuracy scores: %s' % scores)
 
print('Cross Validation accuracy: %.3f +/- %.3f' % (np.mean(scores),np.std(scores)))
```
